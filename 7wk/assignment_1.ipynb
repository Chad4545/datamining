# V2, V3 동시 변환후 Randomization - Full model 결과

- 최종적으로 V2에는 absolute 변환 * sqrt 변환을 해주었으며, V3에는 square 변환을 한 후 Randomization 방법으로 다시 적합정도를 측정하였다.

<img src="pic/ddddd.png">

- Randomization을 위해 100번 반복하였으며, 각 방법당 최소값과 최대값의 차이를 살펴보았다.

      Full model: min = 10.89694, max = 13.72777, mean = 12.45476
      adj       : min = 10.87117, max = 13.67385, mean = 12.45962
      ind       : min = 10.88413, max = 13.65871, mean = 12.57344
      cv        : min = 10.84194, max = 13.7172, mean = 12.48495
      Transform : min = 9.765587, max = 11.89407, mean = 10.94398
      
- 이상치는 없어 보이며, 100번 시행의 평균을 모델의 적합측도로 사용하기로 한다.


### 결론

- Randomization(100번 반복)한 결과 Transformation을 한 Full model의 적합정도가 다른 방법에 비해 월등히 좋다.
