# Stepwise Selection

- Stepwise 방법은 기본적으로 Forward Selection(FS)이라 할 수 있으나, 추가적인 조건으로써 매 단계에서 Backward Elimination(BE)에서와 같이 이미 모형에 선택되어 있는 한 설명변수의 제거가능성을 배제하지 않는다는 점이 다르다. 

- 따라서 Stepwise 방법에서는 앞 단계에서 이루어진 선택에 따라 회귀식에 일단 포함된 설명변수라도 나중 단계에서 제거될 가능성이 있는 것이다.

- 설명변수의 도입이나 제거에 관련된 처리절차는 FS와 BE의 경우와 동일하다.

- 떄떄로 한 설명변수를 모형에 포함시키느냐, 혹은 제거하느냐를 결정하는데 서로 다른 유의수준이 사용될 때도 있다.

- 세부적인 단계는 다음과 같다

---

## 1단계)

- 우선 축소모형이 상수항만을 포함하고 설명변수들은 포함하지 않는다고 가정한다.

---

## 2단계)

- 종속변수 $y$를 설명변수들 $x_1, x_2, ..., x_p$ 각각에 대한 단순선형회귀모형에 적합시킨다.

- 이 단순선형회귀모형들 중에서 결정계수를 가장 크게하는 설명변수를 선형회귀모형에 추가한다.

- 편의상 이 설명변수를 $x_1$이라고 하자.

- 설명변수 $x_1$에 해당하는 회귀계수추정값이 유의적으로 0과 같은 가를 검정한다.

- 만일 귀무가설 $H_0 : B_1 = 0$이 채택되면, 종속변수 $y$에 유의적인 영향을 미치는 설명변수가 없으므로 상수항만 갖는 축소모형을 최적의 선형회귀모형으로 선택한다.

- 만일 이 귀무가설이 기각되면, 제3단계로 넘어간다.

---

## 3단계)

- 제2단계에서 선택한 설명변수 $x_1$과 나머지 $(p-1)$개의 설명변수들 중에서 하나를 짝지어 설명변수들이 2개인 축소모형들을 최소제곱추정한다.

- 즉, 각$i(i=2,3,4,...,p)$에 대해서 설명변수들이 $x_1$과 $x_i$인 축소모형을 최소제곱추정하고, 이 중에서 결정계수가 가장 큰 축소모형을 선택한다.

- 편의상 이 축소모형의 설명변수들을 $x_1$과 $x_2$라고 하자. 

- 설명변수 $x_2$의 회귀계수추정값이 유의적으로 0과 같은 가를 검정한다. 귀무가설 $H_0 : B_2 = 0$를 검정하기 위해서는 축차 F검정통계량을 사용한다. 

- 만일 이 축차F검정통계량값이 임계값보다 작으면, 귀무가설 $H_0 : B_2 = 0 $를 채택하고 설명변수 $x_1$만을 사용하는 선형회귀모형을 최적의 축소모형으로 선택한다.

- 만일 이 축차F검정통계량값이 임계값보다 크면, 부분 F검정법을 사용해서 이미 축소모형에 들어가 있던 설명변수 $x_1$의 회귀계수추정값이 유의적으로 0과 같은가를 검정한다.

- 만일 이 회귀계수추정값이 유의적으로 0과 다르지 않으면, 설명변수 $x_1$을 선형회귀모형에서 제거한다. 

- 따라서 이 경우에는 설명변수 $x_2$만을 사용하는 단순선형회귀모형이 최적의 축소모형으로 선택된다. 

- 만일 설명변수 $x_1$의 회귀계수추정값이 유의적으로 0과 다르면, 제4단계로 넘어간다.

---

## 4단계)

- 제3단계에서 선택한 설명변수들 $x_1, x_2$와 나머지 $(p-2)$개의 설명변수들 중에서 하나를 짝지어 설명변수들이 3개인 축소모형들을 최소제곱 추정한다.

- 즉, 각$i(i=3,4,...,p)$에 대해서 설명변수들이 $x_1, x_2$과 $x_i$인 축소모형을 최소제곱추정하고, 이 중에서 결정계수가 가장 큰 축소모형을 선택한다.

- 편의상 이 축소모형의 설명변수들을 $x_1, x_2, x_3$라고 하자. 

- 설명변수 $x_3$의 회귀계수추정값이 유의적으로 0과 같은 가를 검정한다. 귀무가설 $H_0 : B_3 = 0$를 검정하기 위해서는 축차 F검정통계량을 사용한다. 

- 만일 이 축차F검정통계량값이 임계값보다 작으면, 귀무가설 $H_0 : B_3 = 0 $를 채택하고 설명변수 $x_1, x_2$만을 사용하는 선형회귀모형을 최적의 축소모형으로 선택한다.

- 만일 귀무가설 $H_0 : B_3 = 0 $이 기각되면, 축차 F검정법을 사용해서 이미 축소모형에 들어가 있던 설명변수들 $x_1, x_2$ 각각의 회귀계수추정값이 0과 유의적으로 다른 가를 검정한다.

- 만일 귀무가설 $H_0 : B_1=0$이 채택되면, 선형회귀모형에서 $x_1$을 제거한다. 즉, 설명변수들이 $x_2, x_3$인 선형회귀모형을 최적의 축소모형으로 선택하고 제5단계로 넘어간다.

- 만일 귀무가설 $H_0 : B_2=0$이 채택되면, 선형회귀모형에서 $x_2$을 제거한다. 즉, 설명변수들이 $x_1, x_3$인 선형회귀모형을 최적의 축소모형으로 선택하고 제5단계로 넘어간다.


- 만일 귀무가설 $H_0 : B_1=0, H_0 : B_2=0$을 모두 기각하면, 설명변수들이 $x_1, x_2, x_3$인 선형회귀모형을 최적의 축소모형으로 선택하고 제5단계로 넘어간다.


---

## 5단계)

- 제4단계에서 제5단계로 넘어오는 축소모형들은 설명변수들이 $x_1, x_3$인 경우 , $x_2, x_3$인 경우, $x_1, x_2, x_3$인 경우이다.

- 이 단계에서는 각 축소모형에 새로운 설명변수 하나를 추가한 뒤 제4단계를 반복적용한다.

- 새로이 추가되는 설명변수의 회귀계수추정값이 유의적으로 0과 다르지 않을 때까지 이 단계를 반복한다.
